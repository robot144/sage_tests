# Mechanics example

Variational approach with Euler-Lagrange equation.

We start with a very simple example $m d^z/dt^2 = - g m$. 

Alternatively we can start from the Lagrangian $L(z,\dot{z})=g m z - 1/2 m (\dot{z})^2$.
Consider the change to $J=\int L dt$ from a small change in $z(t)$. We substitute $z=\overline{z}+\epsilon \Delta z$. We assume $\Delta z$ is zero at the edges of the integration. 
Next, we then compute $dJ/d\epsilon$.

$dJ/d\epsilon = d/d\epsilon \int g m z - 1/2 m (\dot{z})^2 dt =  \int g m dz/d\epsilon -  m (\dot{z})d(\dot{z})/d\epsilon dt = \int g m\Delta z -  m (\dot{z})d(\dot{z})/d\epsilon dt$

$= \int g m \Delta z -  m (\dot{z})d(\dot{\Delta z}) dt = \int g m \Delta z dt + \int m (d^2 z/dt^2)\Delta z dt = \int \left( g  m + m  d^2 z/dt^2 \right) \Delta z dt$ 

Next we subsitute linear elements with nodes $t_1=0, t_2=\Delta t, \ldots, t_{K+1}=T$.

$J \approx 1/2 g m \Delta t (z_1+2 z_2+\ldots+2 z_K+Z_{k+1} + 1/2 m /\Delta t( (z_2-z_1)^2 + \ldots + (z_{K+1}-z_K)^2 )$

```julia
using Plots
```

```julia
T=1
K=5
dt=T/K
g=10.00
m=1.0
z=zeros(K+1)
s=0.1 #step size
N=30 #number of iterations
```

```julia
for i=1:N
    grad_z = zeros(size(z))
    for k=2:K
        grad_z[k] = g*m*dt - m/dt*((z[k]-z[k-1])-(z[k+1]-z[k]))
    end
    z=z.+s*grad_z
end
```

```julia
t=0.0:dt:T
print("z_peak=",0.5*g*(0.5*T)^2)
plot(t,z)
plot!(t,0.5*g*((0.5*T)^2 .- (t.-0.5*T).^2))
```
